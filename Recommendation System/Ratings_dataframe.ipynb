```python
import numpy as np
import pandas as pd
import random
```

```python
data = pd.read_csv(r'E:\Jupyterfiles\ML_practice\news_articles.csv')
print(data.shape)
data.head()
```

```python
total_length = 4831 # dataset has 4831 rows
```

# Article Rank

```python
article_rank = []
i = 1
for ar in range(total_length):
    article_rank.append(i)
    i += 1
    
    if i>10:
        i = 1
len(article_rank)
```

```python
data['Article Rank'] = article_rank
```

# Click status

```python
click_session = []
for c in range(total_length):
    click_session.append(random.choice([True, False]))
    
len(click_session)
```

```python
data['Click'] = click_session
```

# Time taken

```python
time_session = []
for t in range(total_length):
    time_session.append(np.random.randint(10, 100))
    
# if click is false set time to 0

for i in range(total_length):
    if click_session[i] is True:
        time_session[i] = time_session[i]
    else:
        time_session[i] = 0

len(time_session)
```

```python
data['Time Spent (seconds)'] = time_session
```

```python
for i in range(total_length):
    if time_session[i] == 0:
        data = data.drop(i)
        total_length -= 1
print(total_length)
```
