```python
# while making an LSTM we need to import model; add input, embedding, dense, dropout and activation layers
from tensorflow.keras import initializers, regularizers, constraints, optimizers, layers
```

```python
from tensorflow.keras.layers import Input, Embedding, Dense, LSTM, Dropout, Activation, GlobalMaxPool1D
from tensorflow.keras.models import Model
```

```python
# imput - lets fix length at 150
inp = Input(shape = (150, ))
```

```python
# embedding - embedding size 75
emb = Embedding(10000, 75)(inp) # num_words initialised at the beginning did not work so manually put 10000
```

```python
# LSTM - in majority examples I went through the output dimension was 60. I am continuing with it
lstm = LSTM(60, return_sequences = True, name = 'lstm_layer')(emb) # sequences and not sequence!!
```

```python
# reduce dimensionality by using Global Max Pooling layer
#from tensorflow.keras.layers import GlobalMaxPool1D
gmp = GlobalMaxPool1D()(lstm)
```

```python
# Droupout (handle missing data) - Fixed value of 0.1
Dout = Dropout(0.1)(gmp)
```

```python
# Activation
# Dense layer
act = Dense(50, activation = "relu")(Dout) # rectified linear unit
```

```python
# send it back to Droupout and then to sigmnoid
Dout2 = Dropout(0.1)(act)
sig = Dense(6, activation = "sigmoid")(Dout2) # here 6 is the number of labels
```

```python
model = Model(inputs = inp, outputs = sig) # input's'
```
