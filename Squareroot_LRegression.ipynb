```python
import numpy as n
import matplotlib.pyplot as plt
```

```python
size = 1000
# vectorization of code
x = n.random.rand(size)
y = n.sqrt(x)
```

```python
plt.figure(figsize = (16, 8))
plt.scatter(
    x,
    y,
    c='black'
)
plt.xlabel("Number")
plt.ylabel("Square root")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(n.array(x).reshape(len(x), 1), n.array(y))
print("The linear model is : sqrt = {:.5} + {:.5}(x)".format(reg.intercept_,reg.coef_[0]))
```
